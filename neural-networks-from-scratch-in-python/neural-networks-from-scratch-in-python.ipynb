# Adding Layers

```python
import numpy as np
import seaborn as sns
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

# sample data
X = np.array([[1, 2, 3, 2.5],
              [2, 5, -1, 2],
              [-1.5, 2.7, 3.3, -0.8]])
```

```python
# data visualization
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(X[0], X[1], X[2], 'pink')
ax.scatter3D(X[0], X[1], X[2], c=X[1])
plt.show()
```

```python
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        self.weights = 0.10 * np.random.randn(n_inputs, n_neurons) # keep initial weights small
        self.biases = np.zeros((1, n_neurons)) # make this non zero if dead network
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
```

```python
n_neurons = 5 # number of neurons in the hidden layer 
n_output = 10 # output size
```

```python
layer1 = Layer_Dense(X.shape[1], n_neurons)
layer2 = Layer_Dense(n_neurons, n_output)
```

```python
layer1.forward(X)
layer2.forward(layer1.output)
```
